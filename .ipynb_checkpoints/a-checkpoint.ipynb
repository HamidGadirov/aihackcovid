```python
import pandas as pd
import numpy as np
owid = pd.read_csv('./covid-19-data-master/public/data/owid-covid-data.csv', index_col='continent')
import seaborn as sns
import matplotlib.pyplot as plt
owid
```

```python
df = owid.sort_values('date', ascending=True)
df['date'] = pd.to_datetime(df['date'], format = '%Y-%m-%d')
owid.tail(20)
```

```python
df_NL = df[df.iso_code=='NLD']
df_HUN = df[df.iso_code=='HUN']
df_IND = df[df.iso_code=='IND']

df_NL_new_tests_smoothed = df_NL['new_tests_smoothed'].dropna()
df_HUN_new_cases_per_million = df_HUN['new_tests_smoothed'].dropna()

df_NL_new_cases_per_million = df_NL['new_cases_per_million'].dropna()
df_HUN_new_cases_per_million = df_HUN['new_cases_per_million'].dropna()

print(len(df_NL_new_tests_smoothed)) # 449
print(len(df_HUN_new_tests_smoothed)) # 466
print(len(df_HUN_new_cases_per_million)) #478

print(np.corrcoef(df_NL_new_tests_smoothed, df_HUN_new_tests_smoothed[17:]))

print(np.corrcoef(df_HUN_new_cases_per_million[12:], df_HUN_new_tests_smoothed))
```

```python
df_NLD = df[df.iso_code=='NLD']
df_IND = df[df.iso_code=='IND']
df_USA = df[df.iso_code=='USA']
df_BRA = df[df.iso_code=='BRA']
df_NER = df[df.iso_code=='NER']

df_NLD_new_cases_per_million = df_NLD['new_cases_per_million'].dropna()
df_IND_new_cases_per_million = df_IND['new_cases_per_million'].dropna()
df_USA_new_cases_per_million = df_USA['new_cases_per_million'].dropna()

# print(len(df_NLD_new_cases_per_million)) # 484
# print(len(df_USA_new_cases_per_million)) # 519
# print(len(df_IND_new_cases_per_million)) # 512

df_NLD_new_cases_per_million = df_NLD_new_cases_per_million[6:]
df_IND_new_cases_per_million = df_USA_new_cases_per_million[41:]
df_USA_new_cases_per_million = df_USA_new_cases_per_million[512-478:]

df_NLD_stringency_index = df_NLD['stringency_index'].dropna()
df_IND_stringency_index = df_IND['stringency_index'].dropna()
df_USA_stringency_index = df_USA['stringency_index'].dropna()

# print(len(df_NLD_stringency_index)) # 478
# print(len(df_USA_stringency_index)) # 509
# print(len(df_IND_stringency_index)) # 478

df_USA_stringency_index = df_USA_stringency_index[31:]

df_NLD_population_density = df_NLD['population_density'].dropna()
df_IND_population_density = df_IND['population_density'].dropna()
df_USA_population_density = df_USA['population_density'].dropna()

# print(len(df_NLD_population_density)) # 484
# print(len(df_USA_population_density)) # 512
print(len(df_IND_population_density)) # 520

df_NLD_population_density = df_NLD_population_density[6:]
df_IND_population_density = df_IND_population_density[512-474:]
df_USA_population_density = df_USA_population_density[42:]
```

```python
min_len_new_cases_per_million = min(len(df_NLD_new_cases_per_million), len(df_USA_new_cases_per_million))
# print(min_len_new_cases_per_million)

min_len_new_cases_per_million = min(len(df_NLD_stringency_index), len(df_USA_stringency_index))

# print("Corr NLD_new_cases_per_million & stringency_index: \n", np.corrcoef(df_NLD_new_cases_per_million, df_NLD_stringency_index))

print("Corr NLD_new_cases_per_million & USA_new_cases_per_million: \n",
      np.corrcoef(df_NLD_new_cases_per_million, df_IND_new_cases_per_million, df_USA_new_cases_per_million))

corr_NLD_USA_new_cases_per_million = np.corrcoef(df_NLD_new_cases_per_million, df_USA_new_cases_per_million)[0][1]
print(corr_NLD_USA_new_cases_per_million)
```

```python
print("Corr NLD_stringency_index & USA_stringency_index: \n", np.corrcoef(df_NLD_stringency_index, df_USA_stringency_index))
corr_NLD_USA_stringency_index = np.corrcoef(df_NLD_stringency_index, df_USA_stringency_index)[0][1]
```

```python
# print(df_USA_population_density)

print("Corr NLD_population_density & USA_population_density: \n", np.corrcoef(df_NLD_population_density, df_USA_population_density))
```
