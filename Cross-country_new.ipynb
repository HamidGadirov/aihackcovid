```python
pd.set_option('display.max_columns', None)

# df_indexed = pd.read_csv('./covid-19-data-master/public/data/owid-covid-data.csv', index_col='continent')
df_indexed = pd.read_csv('./covid-19-data-master/public/data/owid-covid-data.csv')

# Select latest available data
# df_indexed = df_indexed.loc[df_indexed['date'] == '2021-04-01']
df_indexed = df_indexed[df_indexed['date'] > '2021-04-01']

# NLD
# IND
# USA
# BRA
# NER

df_indexed = df_indexed[ (df_indexed.iso_code=='NLD') | (df_indexed.iso_code=='USA') | \
                         (df_indexed.iso_code=='IND') | (df_indexed.iso_code=='BRA') | \
                         (df_indexed.iso_code=='NER') ]
# df_indexed = df_indexed + df_indexed[df_indexed.iso_code=='USA']
# print(df_indexed)

countries = df_indexed['iso_code']
# print(countries)
# countries = df_indexed[df_indexed.iso_code=='NLD'] + df_indexed[df_indexed.iso_code=='USA']
# countries = df_indexed[(df_indexed.iso_code=='NLD') & (df_indexed.iso_code=='IND')]
# data_europe = data_indexed.loc['Europe']

df_indexed_NLD = df_indexed[ (df_indexed.iso_code=='NLD') ]
new_cases_per_million_NLD = df_indexed_NLD['new_cases_per_million']

df_indexed_IND = df_indexed[ (df_indexed.iso_code=='IND') ]
new_cases_per_million_IND = df_indexed_IND['new_cases_per_million']

df_indexed_USA = df_indexed[ (df_indexed.iso_code=='USA') ]
new_cases_per_million_USA = df_indexed_USA['new_cases_per_million']

df_indexed_BRA = df_indexed[ (df_indexed.iso_code=='BRA') ]
new_cases_per_million_BRA = df_indexed_BRA['new_cases_per_million']

df_indexed_NER = df_indexed[ (df_indexed.iso_code=='NER') ]
new_cases_per_million_NER = df_indexed_NER['new_cases_per_million']

df_cases_ = pd.DataFrame({
                       'new_cases_per_million_NLD': new_cases_per_million_NLD.to_numpy(),
                        'new_cases_per_million_IND': new_cases_per_million_IND.to_numpy(),
                        'new_cases_per_million_USA': new_cases_per_million_USA.to_numpy(),
                        'new_cases_per_million_BRA': new_cases_per_million_BRA.to_numpy(),
                        'new_cases_per_million_NER': new_cases_per_million_NER.to_numpy(),
                             })

df_cases_ = df_cases_.reset_index(drop=True)

# df_cases_.set_index(["country"], inplace = True,
#                     append = True, drop = True)

# print(df_cases_.country.unique())
# df_cases_
corr = df_cases_.corr()
corr.style.background_gradient(cmap='coolwarm')
```
